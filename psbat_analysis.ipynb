```python
# Data science imports
import pandas as pd
import numpy as np

# CSV reader
import csv
```

```python
psbat = pd.read_csv("https://raw.githubusercontent.com/chadwickbureau/baseballdatabank/master/core/BattingPost.csv", sep=',')
psbat
```

```python
psbat.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
psbat.columns
```

```python
psbatarr = psbat.to_numpy()
psbatarr
```

```python
col_names = ['yearID', 'round', 'playerID', 'teamID', 'lgID', 'G', 'AB', 'R', 'H',
       '2B', '3B', 'HR', 'RBI', 'SB', 'CS', 'BB', 'SO', 'IBB', 'HBP', 'SH',
       'SF', 'GIDP']
sando_psstats = pd.DataFrame(columns=col_names)
count = 0
```

```python
for a in psbatarr:
    if a[2] == 'sandopa01':
        sando_psstats.loc[count] = a
        count += 1
```

```python
sando_psstats
```

```python
plt.bar(sando_psstats.yearID,  sando_psstats.G)
plt.xlabel("Postseason Year")
plt.ylabel("RBIs")
plt.show()
```

```python
plt.hist2d(sando_psstats.yearID, sando_psstats.HR)
plt.hist2d(sando_psstats.yearID, sando_psstats.RBI)
plt.hist2d(sando_psstats.yearID, sando_psstats.H)
plt.xlabel("Postseason Year")
plt.ylabel("Quantity")
plt.title("Pablo Sandoval Postseason Batting")
plt.show()
```
