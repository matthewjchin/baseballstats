```python
plt.scatter(pitch_stats.yearID, pitch_stats.ERA)
plt.xlabel("Season")
plt.ylabel("ERA")
plt.title("Pitching Earned Run Average, 1871-2020")
plt.show()
```

```python
plt.scatter(pitch_stats.yearID, pitch_stats.W)
plt.xlabel("Season")
plt.ylabel("Wins")
plt.title("Pitching Wins in Season, 1871-2020")
plt.show()
```

```python
pitcher_arr = pitch_stats.to_numpy()
```

```python
scherzer_stats = pd.DataFrame(columns=['playerID', 'yearID', 'stint', 'teamID', 'lgID', 'W', 'L', 'G', 'GS',
       'CG', 'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'SO', 'BAOpp',
       'ERA', 'IBB', 'WP', 'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF', 'GIDP'])
counter = 0


for a in pitcher_arr:
    if a[0] == 'scherma01':
        scherzer_stats.loc[counter] = a
        print(a)
        counter += 1

# scherzer_stats
```

```python
# print(scherzer_stats)
scherzer_stats
```

```python
# Max Scherzer Wins by season

plt.bar(scherzer_stats.yearID, scherzer_stats.W)
plt.xlabel("Season")
plt.ylabel("Wins")
plt.title("Max Scherzer Wins by Season, 2008-2020")
plt.show()
```

```python
# Max Scherzer Win percentage by season

plt.bar(scherzer_stats.yearID, scherzer_stats.W/scherzer_stats.G)
plt.xlabel("Season")
plt.ylabel("Win Percentage")
plt.title("Max Scherzer Win Percentage by Season, 2008-2020")
plt.show()
```
