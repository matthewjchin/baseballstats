```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pitch_stats = pd.read_csv("https://raw.githubusercontent.com/chadwickbureau/baseballdatabank/master/core/Pitching.csv",sep=',')
```

```python
# Return the columns' names
pitch_stats.columns
```

```python
plt.scatter(pitch_stats.yearID, pitch_stats.ERA)
plt.xlabel("Season")
plt.ylabel("ERA")
plt.title("Pitching Earned Run Average, 1871-2020")
plt.show()
```

```python
plt.scatter(pitch_stats.yearID, pitch_stats.W)
plt.xlabel("Season")
plt.ylabel("Wins")
plt.title("Pitching Wins in Season, 1871-2020")
plt.show()
```

```python
pitcher_arr = pitch_stats.to_numpy()
```

```python
scherzer_stats = pd.DataFrame(columns=['playerID', 'yearID', 'stint', 'teamID', 'lgID', 'W', 'L', 'G', 'GS',
       'CG', 'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'SO', 'BAOpp',
       'ERA', 'IBB', 'WP', 'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF', 'GIDP'])
counter = 0


for a in pitcher_arr:
    if a[0] == 'scherma01':
        scherzer_stats.loc[counter] = a
#         print(a)
        counter += 1

scherzer_stats
```

```python
# Max Scherzer Wins by season

plt.bar(scherzer_stats.yearID, scherzer_stats.W, color="navy")
plt.xlabel("Season")
plt.ylabel("Wins")
plt.title("Max Scherzer Wins by Season, 2008-2020")
plt.show()
```

```python
# Max Scherzer Win percentage by season

plt.plot(scherzer_stats.yearID, scherzer_stats.W/scherzer_stats.G, color="navy")
plt.xlabel("Season")
plt.ylabel("Win Percentage")
plt.title("Max Scherzer Win Percentage by Season, 2008-2020")
plt.show()
```

```python
# Sandy Koufax numbers - koufasa01

koufax_stats = pd.DataFrame(columns=['playerID', 'yearID', 'stint', 'teamID', 'lgID', 'W', 'L', 'G', 'GS',
       'CG', 'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'SO', 'BAOpp',
       'ERA', 'IBB', 'WP', 'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF', 'GIDP'])
counter = 0


for a in pitcher_arr:
    if a[0] == 'koufasa01':
        koufax_stats.loc[counter] = a
        counter += 1
koufax_stats
```

```python
plt.bar(koufax_stats.yearID, koufax_stats.W,color="blue")
plt.xlabel("Season")
plt.ylabel("Wins")
plt.title("Sandy Koufax Wins by Season")
plt.show()
```

```python
plt.plot(koufax_stats.yearID, koufax_stats.ERA,color="blue")
plt.xlabel("Season")
plt.ylabel("ERA")
plt.title("Sandy Koufax ERA by Season")
plt.show()
```

```python
# Nolan Ryan numbers - ryanno01

ryan_stats = pd.DataFrame(columns=['playerID', 'yearID', 'stint', 'teamID', 'lgID', 'W', 'L', 'G', 'GS',
       'CG', 'SHO', 'SV', 'IPouts', 'H', 'ER', 'HR', 'BB', 'SO', 'BAOpp',
       'ERA', 'IBB', 'WP', 'HBP', 'BK', 'BFP', 'GF', 'R', 'SH', 'SF', 'GIDP'])
counter = 0


for a in pitcher_arr:
    if a[0] == 'ryanno01':
        ryan_stats.loc[counter] = a
        counter += 1
ryan_stats
```

```python
plt.bar(ryan_stats.yearID, ryan_stats.ERA, color="grey")
plt.xlabel("Season")
plt.ylabel("ERA")
plt.title("Nolan Ryan ERA by Season")
plt.show()
```

```python
plt.bar(ryan_stats.yearID, ryan_stats.W, color="red")
plt.xlabel("Season")
plt.ylabel("Wins")
plt.title("Nolan Ryan Wins by Season")
plt.show()
```

```python
plt.bar(ryan_stats.yearID, ryan_stats.W/ryan_stats.L, color="blue")
plt.xlabel("Season")
plt.ylabel("Win Percentage")
plt.title("Nolan Ryan Percentage by Season")
plt.show()
```
